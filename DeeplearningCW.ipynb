```python
import os
import shutil

# Paths
raw_path = "tiny-imagenet-200"
save_path = "./saved_datasets/tiny_imagenet_images"
train_save_path = os.path.join(save_path, "train")
val_save_path = os.path.join(save_path, "val")

# Create dirs
os.makedirs(train_save_path, exist_ok=True)
os.makedirs(val_save_path, exist_ok=True)

# --- Process Train ---
for class_id in os.listdir(os.path.join(raw_path, "train")):
    class_dir = os.path.join(raw_path, "train", class_id, "images")
    target_dir = os.path.join(train_save_path, class_id)
    os.makedirs(target_dir, exist_ok=True)
    for file in os.listdir(class_dir):
        shutil.copy(os.path.join(class_dir, file), os.path.join(target_dir, file))

# --- Process Val ---
val_annotations = os.path.join(raw_path, "val", "val_annotations.txt")
with open(val_annotations, 'r') as f:
    lines = f.readlines()

for line in lines:
    img_file, class_id = line.strip().split('\t')[:2]
    src = os.path.join(raw_path, "val", "images", img_file)
    dst_dir = os.path.join(val_save_path, class_id)
    os.makedirs(dst_dir, exist_ok=True)
    shutil.copy(src, os.path.join(dst_dir, img_file))

print("Tiny ImageNet images saved successfully in folder format!")
```

```python

# Define paths
save_path = "./saved_datasets/cifar10_images"
train_path = os.path.join(save_path, "train")
test_path = os.path.join(save_path, "test")

# Create directories
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Load CIFAR-10 using torchvision
cifar10_train = CIFAR10(root="./saved_datasets", train=True, download=True)
cifar10_test = CIFAR10(root="./saved_datasets", train=False, download=True)

# Function to save images
def save_cifar_images(dataset, path):
    class_names = dataset.classes  # Class labels
    for class_name in class_names:
        os.makedirs(os.path.join(path, class_name), exist_ok=True)

    for idx, (image, label) in enumerate(dataset):
        class_name = class_names[label]
        image_path = os.path.join(path, class_name, f"{idx}.png")
        image.save(image_path)

# Convert & save images
save_cifar_images(cifar10_train, train_path)
save_cifar_images(cifar10_test, test_path)

print("CIFAR-10 images saved successfully!")

```

```python
import os
import shutil

def filter_first_n_classes(source_dir, target_dir, n_classes):
    """
    Copies the first N sorted class folders from source_dir to target_dir.

    Args:
        source_dir (str): Path to the full dataset (e.g., test split).
        target_dir (str): Where to save the filtered dataset.
        n_classes (int): Number of class folders to copy.
    """
    os.makedirs(target_dir, exist_ok=True)

    # Get all folder names and sort them alphabetically
    class_folders = sorted([d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d))])

    # Select first N classes
    selected_classes = class_folders[:n_classes]

    for class_name in selected_classes:
        src_path = os.path.join(source_dir, class_name)
        dst_path = os.path.join(target_dir, class_name)

        shutil.copytree(src_path, dst_path)
        print(f"Copied: {class_name}")


source_test_dir = '/content/cifar100_images/test'
target_test_dir = '/content/cifar100_subset/test'
num_classes_to_copy = 20

filter_first_n_classes(source_test_dir, target_test_dir, num_classes_to_copy)
```

```python
import os
import shutil
from glob import glob

def merge_datasets(cifar10_path, cifar100_path, save_path):
    os.makedirs(save_path, exist_ok=True)

    # Sort class folders alphabetically for consistent label ordering
    cifar10_classes = sorted(os.listdir(cifar10_path))
    cifar100_classes = sorted(os.listdir(cifar100_path))

    # Map each class to new label: 0–9 for CIFAR-10, 10–109 for CIFAR-100
    combined_classes = {}
    for idx, class_name in enumerate(cifar10_classes):
        combined_classes[class_name] = idx
    for idx, class_name in enumerate(cifar100_classes):
        combined_classes[class_name] = idx + len(cifar10_classes)

    # Copy images with adjusted folder names
    def copy_dataset(source_path, offset=0):
        class_names = sorted(os.listdir(source_path))
        for i, class_name in enumerate(class_names):
            new_label = i + offset
            class_dir = os.path.join(source_path, class_name)
            target_dir = os.path.join(save_path, f"{new_label:03d}_{class_name}")
            os.makedirs(target_dir, exist_ok=True)

            for image_file in glob(os.path.join(class_dir, "*")):
                shutil.copy(image_file, target_dir)

    print("Copying CIFAR-10...")
    copy_dataset(cifar10_path, offset=0)

    print("Copying CIFAR-100...")
    copy_dataset(cifar100_path, offset=len(cifar10_classes))

    print("✅ Merge Complete! Combined dataset saved at:", save_path)
```
